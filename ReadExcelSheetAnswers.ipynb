```python
import pandas as pd
import os
import glob
```

```python
# Path to the folder containing Excel files
folder_path = "/content/drive/MyDrive/Models/answers"

# Get a list of all Excel files in the folder
excel_files = glob.glob(os.path.join(folder_path, "*.xlsx"))

# Dictionary to store the dataframes, with file names as keys
dataframes = []

# Read each Excel file into a dataframe
for excel_file in excel_files:
    # Get the base name of the file without the extension to use as a dictionary key
    file_name = os.path.basename(excel_file).replace(".xlsx", "")

    # Read the Excel file into a dataframe
    dataframe = pd.read_excel(excel_file)

    # Store the dataframe in the dictionary
    dataframes.append(dataframe)
```

```python
dataframes[1].head()
```

```python
answers_counts = {}

for i, df in enumerate(dataframes):
    # Ensure the column "Answers" exists in the dataframe
    if "Answers" in df.columns:
        # Convert the "Answers" column to lowercase
        df["Answers"] = df["Answers"].str.lower()

        # Count the number of each unique element in "Answers"
        counts = df["Answers"].value_counts()

        # Store the counts in the dictionary, using the index as the key
        answers_counts[f"Dataframe {i + 1}"] = counts
    else:
        print(f"'Answers' column not found in dataframe {i + 1}")
```
